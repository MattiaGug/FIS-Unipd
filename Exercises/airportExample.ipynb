```python
import sqlite3

# connect
conn = sqlite3.connect("./exercise_sqlite.db")

# create cursor
cur = conn.cursor()

# define query
all_airports = "SELECT * FROM airports"

```

```python
# execute query (with the cursor)
cur.execute(all_airports)

# get the results (list of tuples)
rows = cur.fetchall()

#print(rows)
```

```python
rows_df = pd.DataFrame(rows)
```

```python
rows_df.head()
```

```python
# limit the result to the first three rows

q2 = "SELECT * FROM airports LIMIT 3"

# execute query (with the cursor)
cur.execute(q2)

# get the results (list of tuples)
rows = cur.fetchall()

# create dataframe
rows_df = pd.DataFrame(rows)

# print results
rows_df.head()
```
