### <font color='#e76f51'>Custom python modules</font>

A Python module is essentially a file that contains Python definitions, statements, functions, classes, variables, and even executable code. By organizing related code into a module, you not only enhance the code's clarity and comprehensibility but also create a logical and structured organization for your code.

This is employed to reuse functions in various scripts without the necessity of redefining them each time. Consider a file named `mymodule.py`, containing these two functions

```python
def add(x,y):
    return x+y

def subtract(x,y):
    return x-y
```

Then, on another script (located in the same directory), I can import and use these functions without the need of redefining them. This can be achieved with the statement `import modulename`. In this case, we can use our module in a script by

```python
import mymodule

print(mymodule.add(1,2))
print(mymodule.subtract(10,4))
```

If you are on a notebook, the import statement doesn't need to be run every time you want to execute on cell. It is quite common indeed to find all the imports needed on the first cell of a notebook.

We can use an alias for the module name with the `import` followed by an `as`+the name we want to use

```python
import mymodule as mm

print(mm.add(1,2))
```

It is possible to import individual function from a module using `from`+`import`. In this case we can use directly the function name.

```python
from mymodule import add

print(add(1,2))
```



### <font color='#e76f51'>Built-in python modules</font>

There is a set of Python modules that are "Built-in", e.g. coming together with every python installation.

One example, is the `random` module. It contains functions that can be used to make/work with random numbers.

Similarly to custom module, we can import it with

```python
import random
```

this comes with a large set of functions and utilities ([documentation](https://docs.python.org/3/library/random.html)).

From sampling numbers from distributions, e.g. `random.gauss(mu=0.0, sigma=1.0)` to chosing random elements from a sequence `random.choice(seq)` and shuffling lists.

Similarly, the `math` module can be used to perform mathematical operations.

Another useful module is `time`. You can use it to measure the execution time of a function/delay it

````python
import time

start = time.time() # return the current time

# do nothign for 1s
time.sleep(1)

stop = time.time()

print('Total execution time: {:.4f}s'.format(stop-start))
````

Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes, an area where points can be specified in terms of x-y coordinates

The first type of plot we are exploring is the scatter plot. It takes two collections of items and uses the points from the first collection as the x-coordinates and the points from the second collection as the y-coordinates.

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots() # Create a figure containing a single axes.

x = [1,2,3,4,5,6]
y = [1,5,3,4,5,2]

ax.plot(x,y) # plot data on the current ax
#fig.show() # show the figure (canvas + all axes)
````

We can obviously customize the plot

````python
fig, ax = plt.subplots(figsize=(8,4))

x  = [1,2,3,4,5,6]
y1 = [1,5,3,4,5,2]
y2 = [3,4,2,3,2,5]

ax.plot(x, y1, 'o-', label='data 1')
ax.plot(x, y2, 'o-', label='data 2')

ax.set_xlabel('X-label')
ax.set_ylabel('Y-label')
ax.set_title('Awesome plot')
ax.legend()
````

If we want to show more plots on the same figure it is possible to create a grid of axes. Let's say that we want two plots on the same line

````python
fig, axarr = plt.subplots(1,2,figsize=(8,4))

x  = [1,2,3,4,5,6]
y1 = [1,5,3,4,5,2]
y2 = [3,4,2,3,2,5]

axarr[0].plot(x, y1, 'o-')
axarr[1].plot(x, y2, 'o-')
````

**Exercise**

Write a function to simulate a 1D random walk. To simulate it, use the following algorithm:

Initialize an object at position y=0. Then, for every iteration:
*  Randomly move the object up or down. In other words, add or subtract 1 from the y value. (hint use the random module to pick the direction)
* Keep track of all the positions of the object, that is, the y values.

The function must return two arrays: one with the iteration number and one with the y positions. The input argument is the number of iterations `N`.

Execute the function 4 times and plot the results on the same figure.

````python
import math
import random
import matplotlib.pyplot as plt

def random_walk_1d(N):
    y = 0
    positions = [y]
    iterations = [0]
    
    for i in range(1, N + 1):
        step = random.choice([-1, 1])
        y += step
        positions.append(y)
        iterations.append(i)
    
    return iterations, positions

# Execute the function 4 times and store results
num_iterations = 100
results = []

for i in range(4):
    iterations, positions = random_walk_1d(num_iterations)
    results.append((iterations, positions))

# Plot the results on the same figure
plt.figure(figsize=(10, 6))
plt.xlabel('Iteration')
plt.ylabel('Position (y)')
plt.title('1D Random Walk Simulation')

for i in range(len(results)):
    iterations, positions = results[i]
    plt.plot(iterations, positions)
    
````

### <font color='#e76f51'>(Extra) Exernal modules - Numpy</font>

[NumPy](https://numpy.org/) (have a look at the quickstart tutorial) is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object (numpy.ndarray), various derived objects such as masked arrays and matrices, and an assortment of routines for fast operations on arrays.

It can be installed with
```python
pip install numpy
```

and imported into your script/notebook
```python
import numpy as np
```

The main difference between the ndarray and Python list is that ndarray has fixed size after declaration. You can print its dimension with
```
print(arr.shape)
```
This will print a touple with as many elements as the number of dimensions of the array. I the case of a 2x2 matrix this will be (2,2).

It does not mean that the addition of elements is forbidden, but it does technically require the creation of a new array and the destruction of the starting one.

Therefore, in general you need to first fill a list and then create the numpy array from it
* you can also use methods of numpy to create directly an array

```python
import numpy as np

l = []
for i in range(10):
    l.append(i)

arr = np.array(l)
```

In python, if you are using a loop to fill a list you can use the so-called list-comprehension. To fill a list in the same way as before, you can use
```python
l = [i for i in range(10)]
arr = np.array(l)
```

or, if you want to squeeze everythin in one line
```python
arr = np.array([i for i in range(10)])
```

Some notable buil-in methods used to create python arrays are the following
```python
print(np.zeros(10)) # numpy array with 10 elements set to 0
print(np.arange(10,20, 2)) # array with elements from 10 to 20 at steps of 2
print(np.linspace(0, 2, 9)) # array with 9 numbers from 0 to 2 (equally spaced)
```

Numpy is used in python because it is much faster than working with lists. For example, let's say that we want to create a list and then perform some mathematical operations on it

```python
l = []
for i in range(10):
    l.append(i)
```

later in our script we need to increment every element by one and multiply by 10.
```python
for i in range(len(l)):
    l[i] = (l[i] + 1)*10
```

Let's try to do the same with numpy.

```python
arr = np.array([i for i in range(10)])

#... more code

arr = (arr+1)*10
```

this is possible because the operation has been "broadcasted" to every element of the array and made the for loop implicit.

Multidimensional arrays can be viewed as "Arrays of Arrays". For example, to create a matrix with shape (2,3) we can do this

```python
matrix = np.array([[1,2,3],[4,5,6]])
print(matrix)
```

and the elements can be accessed with
```python
print(matrix[row_num, col_num])
# or
print(matrix[row_num][col_num])
```

It is possible to change the shape of a matrix/array by reshaping it
```python
print(matrix.shape)
matrix_new = matrix.reshape(3,2)
print(matrix_new.shape)
```

Slicing is also available for numpy arrays. For example, if we want to select all the rows of a matrix starting from the first one we can do
```python
print(matrix[1:,:])
```
